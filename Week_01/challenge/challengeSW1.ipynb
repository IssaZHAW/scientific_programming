```python
import numpy as np

numbers = np.random.randn(1000)

numbers
```

```python
quartiles = np.percentile(numbers, [25, 50, 75])
quartiles
```

```python
import matplotlib.pyplot as plt

plt.hist(numbers)
plt.show()
```

```python
plt.boxplot(numbers)
plt.title("Boxplot of Random Numbers")
plt.show()
```
