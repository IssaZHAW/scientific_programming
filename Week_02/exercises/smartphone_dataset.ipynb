```python
import pandas as pd
import numpy as np

# Zufallsgenerator für Reproduzierbarkeit
np.random.seed(42)

# Mögliche Hersteller und Modelle
makes = ['Apple', 'Samsung', 'Google', 'OnePlus', 'Xiaomi', 'Motorola', 'Nokia', 'Sony']
models = {
    'Apple': ['iPhone 15', 'iPhone 15 Pro', 'iPhone 15 Plus'],
    'Samsung': ['Galaxy S24', 'Galaxy A54', 'Galaxy Z Fold 5'],
    'Google': ['Pixel 8', 'Pixel 8 Pro', 'Pixel 8a'],
    'OnePlus': ['12', '12 Pro', '11'],
    'Xiaomi': ['14', '14 Pro', '13T'],
    'Motorola': ['Edge 50', 'G54', 'Razr'],
    'Nokia': ['G310', 'C210', 'X100'],
    'Sony': ['Xperia 1 V', 'Xperia 5 V']
}

# Datensatz generieren
n_phones = 1000
data = {
    'make': [],
    'model': [],
    'price': [],
    'camera_resolution': [],
    'battery_life_hours': [],
    'storage_size': [],
    'screen_size': []
}

for _ in range(n_phones):
    make = np.random.choice(makes)
    model = np.random.choice(models[make])
    
    # Realistische Preisrange basierend auf Hersteller
    price_range = {'Apple': (800, 1500), 'Samsung': (300, 1300), 'Google': (400, 1100),
                   'OnePlus': (400, 900), 'Xiaomi': (250, 800), 'Motorola': (200, 700),
                   'Nokia': (150, 500), 'Sony': (700, 1400)}
    price = np.random.uniform(*price_range[make])
    
    camera_resolution = np.random.choice([12, 13, 48, 50, 64, 108])
    battery_life_hours = np.random.randint(15, 35)
    storage_size = np.random.choice([64, 128, 256, 512, 1024])
    screen_size = np.random.uniform(5.0, 6.9)
    
    data['make'].append(make)
    data['model'].append(model)
    data['price'].append(round(price, 2))
    data['camera_resolution'].append(camera_resolution)
    data['battery_life_hours'].append(battery_life_hours)
    data['storage_size'].append(storage_size)
    data['screen_size'].append(round(screen_size, 2))

# DataFrame erstellen
df = pd.DataFrame(data)

# Zusammenfassung anzeigen
print(f"Datensatz mit {len(df)} Smartphones erstellt:")
print(df.head(10))
print(f"\nDatensatz-Info:\n{df.info()}")
print(f"\nBeschreibende Statistik:\n{df.describe()}")

df.to_excel("smartphones.xlsx", index=False)

print("Excel-Datei 'smartphones.xlsx' wurde erstellt.")
```
